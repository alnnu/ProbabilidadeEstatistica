```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv("dados_estudantes.csv", sep=";")
```

```python
df.head(5)
```

```python
legendas = {
    "Id": {"title": "Indentificação do Aluno", "labels": ""},
    "Turma": {"title": "Turma Atribuida", "labels": ""},

    "Sexo": {
        "title": "",
        "labels":{ 
            "F":"Feminino",
            "M":"Masculino",
            "NI":"Não Informado"
        }
    },

    "Idade": {"title": "Idade em Anos", "labels": ""},
    "Altura": {"title": " Altura em metros", "labels": ""},
    "Peso": {"title": " Peso em Kg", "labels": ""},
    "Filhos": {"title": "Numero de Filhos na Familia", "labels": ""},

    "Fuma":{
        "title": "Hábito de fumar",
         "labels":  {
             "SIM":"sim",
             "NAO":"não"
             }
    },

    "Toler": {
        "title": "Tolerancia ao Cigarro",
        "labels": {
            "I": "Indifernte",
            "P": "Incomoda Pouco",
            "M": "Incomoda Muito"
        }
    },

    "Exerc": {"title": "Horas de atividade Física por Semana", "labels": ""},
    "Cine": {"title": "Numero de vezes que vai ao cinema por semana", "labels": "" },
    "OpCine": {"title":"Opinião a respeito das Salas de Cinema na Cidade", "labels": ""},
    "TV": {"title":"Horas gastas assistindo TV, por semana", "labels": ""},
    
    "OpTV": {
        "title":"Opinião a respeito da qualidade da qualidade da TV",
        "labels":{
            "R": "Ruim",
            "M": "Media",
            "B": "Boa",
            "N": "Não Sabe",
            }
        }
}
```

```python
def plot_grafs(col: str, histogram:bool=False):
    import random
    plt.style.use('_mpl-gallery') 
    data = df[col].value_counts().reset_index()
    tint = ("red", "green", "blue", "yellow", "pink", "purple", "gray", "cyan" )
    colors = dict()
    color = list()

    print(col)
    if col != "Alt" and col != "Peso" and col != "Exerc" and  col != "TV": 
        for index, i in enumerate(data.iloc[:,0]):
            color.append(tint[index])
            colors.update({i: tint[index]})
            labels = list(colors.keys())
            handles = [plt.Rectangle((0,0),1,1, color=colors[label]) for label in labels]
            labels = legendas[col]["labels"]


    if col != "Idade" and col != "Alt" and  col != "Peso" and col != "Exerc" and col != "TV":
        fig, ax = plt.subplots(1,2, figsize=(12,6))
        fig.tight_layout(pad=5.0)

        # PIE PLOT 
        ax[0].pie(data["count"], labels=data[col], autopct="%.2f%%",  pctdistance=1.2, labeldistance=.6, colors=color)
        ax[0].title.set_text(f"Porcentagem de {legendas[col]['title']} ")

        #BAR PLOT 
        ax[1].bar(data[col], data["count"], color=color)

        ax[1].title.set_text(f"Taxa de {col}")
        ax[1].legend(handles, legendas[col]["labels"])

        plt.show()

    elif histogram:
        fig, ax = plt.subplots(figsize=(12,6))
        fig.tight_layout(pad=5.0)

        #HISTOGRAM PLOT
        ax.hist(df[col], color="blue", ec="white")
        ax.set_title(f"Histograma de {col}")
        plt.xscale("linear")
        plt.show()

    else:
        fig, ax = plt.subplots(figsize=(12,6))
        fig.tight_layout(pad=5.0)

        #BAR PLOT
        ax.bar(data[col], data["count"], color="blue", ec="white" )
        ax.set_title(f"Gráfico de {legendas[col]['title']}")

        plt.xticks(np.arange(min(data[col])-1, max(data[col])+1))
        plt.ylabel(f"Quantidade de alunos")
        plt.yscale(value="linear")
        plt.xlabel(f"{col}")
        plt.legend(legendas[col]["labels"])
        # plt.xscale(value="")
        plt.show()
```

```python
plot_grafs("Fuma")
```

```python

fig, ax = plt.subplots()
ax.axis('off')
ax.axis('tight')
t= ax.table(cellText=df.values, colWidths = [0.3]*len(df.columns),  colLabels=df.columns,  loc='center')
t.auto_set_font_size(False) 
t.set_fontsize(8)
fig.tight_layout()
plt.show()
```
